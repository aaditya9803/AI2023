# Correlation

## $$ \rho _{xy} = corr(x,y)$$

## $$ -1.0 \leq \rho_{xy} \leq +1.0 $$

```python
data['PRICE'].corr(data['AveRooms'])
```

```python
data.corr()
```

```python
#Visualising correlation witha Heatmap
mask=np.zeros_like(data.corr())
print(mask)
```

```python
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices]=True
print (mask)
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(data.corr(), mask = mask, annot=True, annot_kws={"size":14})
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
sns.set_style('white')
plt.show()
```

```python
plt.scatter(x=data['PRICE'], y=data['MedInc'])
plt.xlabel('PRICE - Price of the house', fontsize=14)
plt.ylabel('MedInc', fontsize=14)
medinc_corr = round(data['MedInc'].corr(data['PRICE']), 3)
plt.title(f'PRICE v Median Income(correlation {medinc_corr})', fontsize=14)
plt.show()
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('whitegrid')
# sns.jointplot (x=data['PRICE'], y=data['MedInc'], size=7,color='blue', joint_kws={'alpha':0.5})
sns.jointplot (x=data['PRICE'],y=data['MedInc'],color='red',kind='hex')

plt.show()
```

```python
#Pair plots & Jupyter Microbenchmarking Techniques
sns.lmplot(x='PRICE', y='AveRooms', data=data)
plt.show()
```

```python
averooms_corr = round(data['AveRooms'].corr(data['PRICE']), 3)
plt.figsize=(9,6)
plt.scatter(x=data['PRICE'], y=data['AveRooms'], alpha=0.6,s=80,color='skyblue')
plt.xlabel('PRICE - Price of the house', fontsize=14)
plt.ylabel('Average rooms', fontsize=14)
plt.title(f'PRICE v Average rooms(correlation {averooms_corr})', fontsize=14)
plt.show()
```

```python
sns.lmplot(x='AveRooms',y='PRICE',data=data)
plt.show()
```
